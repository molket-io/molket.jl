## A quick example of the use of the MolKet package

```julia
include("../molket.jl")
```

 We can intialize a quantum register with a given number of qubits, also with a specific convention:
big-endian or little-endian.


```julia
qc = qc_init(2)
show_statevector(qc)
```

```julia
qc = qc_init(2,big_endian=true)
show_statevector(qc)
```

```julia
qc = qc_init(2,big_endian=false)
show_statevector(qc)
```

Applying gates: Hadamard gate, Pauli gates, CX gate, ... etc using the tensor properitie

```julia
qc = qc_init(2)
op(qc, q.H, 0)
show_statevector(qc)
```

```julia
qc = qc_init(2)
println("Apply H on the first qubit q_0")
op(qc, q.H, 0)
show_statevector(qc)
println("")
println("Apply H on the second qubit q_1")
op(qc, q.H, 1)
show_statevector(qc)
```

You can also also have a peek at the states of the quantum register at any time.


```julia
peek_states(qc)
```

```julia
qc = qc_init(2)
#show_statevector(qc)
op(qc, q.H, 0)
op(qc, q.H, 1)
show_statevector(qc)
peek_states(qc)
```

```julia
ind_qubit = 0
qc,sv_0,sv_1,p0,p1 = z_measure(qc, 0,big_endian=true)
qc.state_vector = sv_0
show_statevector(qc)
peek_states(qc)
```

```julia
# Here, you can measure the states of the qubits and simulate with a given number of shots 
qc = qc_init(2)
op(qc, q.H, 0)
op(qc, q.H, 1)
show_statevector(qc)
peek_states(qc)
shots = 1000
# statevector simulator
measure_state(qc,shots;save_fig=false)
# the statevector is normalized, check the probabilities of the states.
# Check for yourself that the probabilities are the same as or close to 
# ... the statevector simulator
```
